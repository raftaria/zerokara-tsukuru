```python
# 함수
def hello():
    print('Hello World!')

hello()

def hello(object):
    print('Hello ' + object + '!')

hello('cat')
```

```python
# 클래스
class Man:
    def __init__(self, name):
        self.name = name
        print('Initialized!')

    def hello(self):
        print("Hello " + self.name + '!')
    
    def goodbye(self):
        print('Good bye ' + self.name + '!')

m = Man("David")
m.hello()
m.goodbye()
```

## 넘파이

```python
import numpy as np

x = np.array([1.0, 2.0, 3.0])
print(x)
print(type(x))
```

```python
x = np.array([1.0, 2.0, 3.0])
y = np.array([2.0, 4.0, 6.0])
print(x + y)
print(x - y)
print(x * y)
print(x / y)
```

```python
# 브로드 캐스트
x = np.array([1.0, 2.0, 3.0])
print(x / 2.0)

A = np.array([[1, 2], [3, 4]])
B = np.array([10, 20])
print(A * B)
```

```python
# 넘파이의 N차원 배열
A = np.array([[1, 2], [3, 4]])
print(A)
print(A.shape)
print(A.dtype)
```

```python
B = np.array([[3, 0], [0, 6]])
print(A + B)
print(A * B)

print(A * 10)
```

```python
# 원소 접근
X = np.array([[51, 55], [14, 19], [0, 4]])
print(X)
print(X[0])
print(X[0][1])

for row in X:
    print(row)

X = X.flatten()
print(X)
print(X[np.array([0, 2, 4])])
```

### Matplotlib

```python
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1)
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
x = np.arange(0, 6, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos")
plt.xlabel("x")
plt.ylabel("y")
plt.title("sin & cos")
plt.legend()
plt.show()
```
